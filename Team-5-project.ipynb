# 2. Data Cleaning and transformation

```python
employee.dropna(axis=0,subset=['Attrition'],inplace=True)
employee.drop(["Application ID","EmployeeNumber"],axis=1, inplace=True)
employee.isnull().sum()
```

```python
def fill_mean_na(mean):
    for mn in mean:
        employee[mn].fillna(employee[mn].mean(),inplace=True)
def fill_mode_na(mode):
    for md in mode:
        employee[md].fillna(employee[md].mode()[0],inplace=True)
def fill_unknown_na(unknown):
    for un in unknown:
        employee[un].fillna("unknown",inplace=True)
```

```python
employee["Age"].fillna(round(employee['Age'].mean()),inplace=True)
employee["EmployeeCount"].fillna(1,inplace=True)
mean = ["DailyRate","DistanceFromHome","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager","PercentSalaryHike","StandardHours","TotalWorkingYears","HourlyRate","MonthlyRate","MonthlyIncome","NumCompaniesWorked"]
fill_mean_na(mean)
mode = ["WorkLifeBalance","PerformanceRating","RelationshipSatisfaction","StockOptionLevel","Education","EnvironmentSatisfaction","EnvironmentSatisfaction","JobInvolvement","JobLevel","JobSatisfaction",]
fill_mode_na(mode)
unknown = ["Attrition","Department","EducationField","Employee Source","JobRole","BusinessTravel","MaritalStatus","Over18","OverTime","Gender"]
fill_unknown_na(unknown)

employee.info()
```

```python
employee.drop_duplicates(inplace =True)
employee_copy = employee.copy()
LE = LabelEncoder()
employee
```

```python
for un in unknown:
    employee_copy[un] = LE.fit_transform(employee_copy[un])
employee_copy.info()
```

# 3. Analysis Questions

What are the top most characterises of the employees who take volunteer resignation?



```python
fig, ax = plt.subplots(figsize=(18,12))   
sns.heatmap(employee_copy.corr(), ax=ax)
```

What are the top most departments and job role of the employees who take volunteer resignation?

```python
emp = employee[employee.Attrition == 'Voluntary Resignation'].groupby(["Department"])["EmployeeCount"].count().nlargest(5)
sns.barplot(x = emp.index, y = emp.values)
```

```python
emp = employee[employee.Attrition == 'Voluntary Resignation'].groupby(["JobRole"])["EmployeeCount"].count().nlargest(5)
sns.barplot(x = emp.index, y = emp.values)
plt.xticks(rotation=90)
```

What number of employees were promoted and still take volunteer exit?

```python
emp = employee[(employee.Attrition == 'Voluntary Resignation') & (employee.YearsSinceLastPromotion != 0)].groupby(["JobRole"])["EmployeeCount"].count()
sns.barplot(x = emp.index, y = emp.values)
plt.xticks(rotation=90)
```

What are the employee source who have taken volunteer exit?

```python
emp = employee[employee.Attrition == 'Voluntary Resignation'].groupby(["Employee Source"])["EmployeeCount"].count()
sns.barplot(x = emp.index, y = emp.values)
plt.xticks(rotation=90)
```

Find how much years employee work at company who have taken volunteer exit?


```python
emp = employee[employee['Attrition'] == "Voluntary Resignation"] 
sns.countplot(x=emp['TotalWorkingYears'])
plt.xticks(rotation=90)
```

What is the performance rating of rating who got fired?

```python
emp = employee[employee['Attrition'] == "Voluntary Resignation"] 
sns.countplot(x=emp['PerformanceRating'])
```

What is the average monthly income of employee who gave voluntary resignations?

```python
emp = employee[employee.Attrition == 'Voluntary Resignation'].groupby(["JobRole"])["MonthlyIncome"].mean()
sns.barplot(x = emp.index, y = emp.values)
plt.xticks(rotation=90)
```

What is the business travel type of employee who gave voluntary resignations?


```python
emp = employee[employee.Attrition == 'Voluntary Resignation'].groupby(["BusinessTravel"])["EmployeeCount"].count().nlargest(5)
sns.barplot(x = emp.index, y = emp.values)
```
